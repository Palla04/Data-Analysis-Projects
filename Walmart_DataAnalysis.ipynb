The df.isnull().sum() method is a very useful way to quickly check for missing values in your DataFrame. Here's a breakdown:

df.isnull(): This part of the code returns a boolean DataFrame of the same shape as df, where True indicates a missing value (NaN, None, NaT) and False indicates a non-missing value.
.sum(): When sum() is called on a boolean DataFrame, it counts the True values (which are treated as 1) and False values (treated as 0) for each column. Therefore, df.isnull().sum() effectively gives you the total count of missing values for each column in the DataFrame.

**Cleaning Logic for Vehicle Type**

If the 'vehicleType' is missing for a row then we have to check the value (X) that is present for 'model' column. Now we have to check the first 'vehicleType' (Y) from the top for the value 'X', now replace null value by that value (Y)

```python
model_grouped_vehicle_type = df.groupby('model')['vehicleType'].transform(
    lambda x: x.dropna().iloc[0]
    if not x.dropna().empty
    else np.nan)

print(model_grouped_vehicle_type)

df['vehicleType'].fillna(model_grouped_vehicle_type, inplace=True)

print("Null values in 'vehicleType' after filling:")
print(df['vehicleType'].isnull().sum())
```

1. For each model, it picks the first Non-Null vehicle type.
2. If all values for that model are null, it returns NaN
3. So if a model has no vehicleType at all, it can not fill anything -> those remain null.

**Cleaning Logic:**

If for a row 'grearbox' is null, then find out the 'brand' (X) in that row. Now from the dataset we have to find the first 'gearbox' value(Y) for that 'brand' (X), now replace null value of gearbox by Y.

```python
model_grouped_gearbox = df.groupby('brand')['gearbox'].transform(
    lambda x: x.dropna().iloc[0]
    if not x.dropna().empty
    else np.nan)

df['gearbox'].fillna(model_grouped_gearbox, inplace=True)

print("Null values in 'gearbox' after filling:")
print(df['gearbox'].isnull().sum())
```

**Find out Brand wise Top Model where model count is higher for that brand.**

```python
top_models_per_brand = df.groupby(['brand', 'model']).size().reset_index(name='model_count')
top_models_per_brand = top_models_per_brand.loc[top_models_per_brand.groupby('brand')['model_count'].idxmax()]
print("Brand-wise Top Model with highest count:")
display(top_models_per_brand.set_index('brand'))
```

Now if model is null, then find its brand name and replace null value of model by top model of that brand

```python
top_model_map = top_models_per_brand.set_index('brand')['model'].to_dict()

df['model'] = df['model'].fillna(df['brand'].map(top_model_map))

print("Null values in 'model' after filling:")
print(df['model'].isnull().sum())
```
